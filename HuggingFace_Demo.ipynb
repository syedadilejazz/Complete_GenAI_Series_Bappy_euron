Sentiment Analysis(Kind of Text Classification)

```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis")#Download default model for sentiment analysis
result=classifier("I was so not happy withthe Last Mission Impossible movie")
print(result)
```

```python
pipeline(task="sentiment-analysis")("I am confused with the Baribie movie")
```

Batch Sentiment Analysis

```python
pipeline(task="sentiment-analysis",model="facebook/bart-large-mnli")\
("everyday lots of llm papers are publisehed about LLMs Evaluation.\
lots of them looks very promising\
I am not sure if we cab actaully Evaluate LLMS.\
There is still lots to do.\
Don't you think?"
)
```

```python
classifier=pipeline(task="sentiment-analysis",model="facebook/bart-large-mnli")
task_list=["I relayy like Autoencoders, best model for Anomaly Detection",\
           "I am not sure if we CAN actualy Evaluate LLMs.",\
           "PassiveAgressive is the name of Linear Regression odel that so many people do not know.",\
           "I hate long Meetings."
           ]
classifier(task_list)
```

```python
text="""Yin et al. proposed a method for using pre-trained NLI models as a ready-made zero-shot sequence classifiers. The method works by posing the sequence to be classified as the NLI premise and to construct a hypothesis from each candidate label. For example, if we want to evaluate whether a sequence belongs to the class "politics", we could construct a hypothesis of This text is about politics.. The probabilities for entailment and contradiction are then converted to label probabilities."""
final_text=[x.strip() for x in text.split(".")]
print(final_text)
```

```python
result=classifier(final_text)
for text,result in zip(final_text,result):
  print(text,"---------",result)
```

```python
classifier=pipeline(task="sentiment-analysis",model="SamLowe/roberta-base-go_emotions")
result=classifier(task_list)
for text,result in zip(task_list,result):
  print(text,"---------",result)


```
