```python
#Decode the text
decoded_output=tokenizer.decode(encoded_inputs["input_ids"])
print(decoded_output)
```

Fine Tuning Pretrained Model

Step1-Import necesssary packages

```python
!pip install datasets
```

Step2-Import dataset

```python
from datasets import load_dataset
dataset=load_dataset('stanfordnlp/imdb')
print(dataset)
```

```python
dataset['train'][0]
```

Step3-Proprocess the Data

Tokenize the dataset using the tokenizer assocaited with the pre-trained Model

```python
from transformers import  AutoTokenizer
tokenizer=AutoTokenizer.from_pretrained('llm-semantic-router/lora_intent_classifier_bert-base-uncased_model')
#Tokenize the dataset
def tokenizer_function(examples):
  return tokenizer(examples["text"],padding="max_length",truncation=True)
tokenized_datasets=dataset.map(tokenizer_function,batched=True)
```

```python
tokenized_datasets
```

```python
tokenized_datasets['train'][0]
```

Ste4.Setup the Training Argumment
Specify the hyperparameters and training settings

```python
from transformers import TrainingArguments
training_args=TrainingArguments(
    output_dir='./results',
    eval_strategy='epoch',
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=1,
    weight_decay=0.01)

training_args
```

Step5:Inittialize the model, Load the pred-trained model abd define the training procedure

```python
from transformers import AutoModelForSequenceClassification,Trainer
#Load pretrained model
model=AutoModelForSequenceClassification.from_pretrained('llm-semantic-router/lora_intent_classifier_bert-base-uncased_model')
#Initialize the trainer
trainer=Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets['train'],
    eval_dataset=tokenized_datasets['test']
)
```

Step6:Train the model,Actual train of the model,Training Loss and Validation Loss will decrease as we increase the epoches

```python
trainer.train()
```

Step7:Evaluate the model,eval_loss should be close the 0


```python
results=trainer.evaluate()
print(result)
```

Step8:Save the Fine-Tuned Model

```python
#save the model
model.save_pretrained('./fine_tuned_model')
tokenizer.save_pretrained('./fine_tuned_model')
```

```python
import arvix
import pandas as pd
```

```python
#Query to fetch
query='ai OR artificial intelligence or Machine Learning'
search=arvix.Search(query=query,max_results=10,sort_by=arvix.SortCriterion.SubmittedDate)
#Fetch papers
for result in search.results():
  papers.append({
      'published':result.published,
      'title':result.title,
        'abstarct':result.summary,
        'authors':result.authors,
        'pdf_url':result.pdf_url})
  #Convert to DataFrame
df=pd.DataFrame(papers)
pd.set_option('display.max_clolwidth',None)
df.head(10)
```

```python
#Example abstarct from API
abstract=df['abstract'][0]

summarize=pipeline('summarization',model="facebook/bart-large-cnn")
summarization_result=summarize(abstract)
print(summarization_result[0]['summary_text'])
```
